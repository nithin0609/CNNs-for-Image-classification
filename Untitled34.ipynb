```python
 import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
import seaborn as sns 

# Load the CIFAR-10 dataset
(x, y), (x_test, y_test)  = keras.datasets.cifar10.load_data()

# Normalize the datasets
x = x / 255.0
x_test = x_test / 255.0

# Split the data into training, validation, and testing datasets
x_train, x_val = x[:35000], x[35000:50000]
y_train, y_val = y[:35000], y[35000:50000]

# Check the class distribution in each dataset
classes_name = ['Airplane', 'Automobile', 'Bird', 'Cat', 'Deer', 'Dog', 'Frog', 'Horse', 'Ship', 'Truck']

# Plotting class distribution in the training set
plt.figure(figsize=(8,6))
classes, counts = np.unique(y_train, return_counts=True)
sns.barplot(classes_name, counts)
plt.title('Class distribution in training set')
plt.tight_layout()
plt.show()

# Plotting class distribution in the validation set
plt.figure(figsize=(8,6))
classes, counts = np.unique(y_val, return_counts=True)
sns.barplot(classes_name, counts)
plt.title('Class distribution in validation set')
plt.tight_layout()
plt.show()

# Plotting class distribution in the testing set
plt.figure(figsize=(8,6))
classes, counts = np.unique(y_test, return_counts=True)
sns.barplot(classes_name, counts)
plt.title('Class distribution in testing set')
plt.tight_layout()
plt.show()
```

# LeNet Model Training

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
from sklearn.metrics import confusion_matrix, accuracy_score

# 1. Data Splitting
x_train_full, x_test, y_train_full, y_test = x_train[:45000], x_train[45000:], y_train[:45000], y_train[45000:]
x_train, x_val, y_train, y_val = x_train_full[:35000], x_train_full[35000:], y_train_full[:35000], y_train_full[35000:]

# 2. Define the LeNet Model
def build_lenet():
    lenet = keras.models.Sequential([
        keras.layers.Conv2D(6, kernel_size=5, strides=1, activation='relu', input_shape=(32,32,3), padding='same'),
        keras.layers.AveragePooling2D(),
        keras.layers.Conv2D(16, kernel_size=5, strides=1, activation='relu', padding='valid'),
        keras.layers.AveragePooling2D(),
        keras.layers.Conv2D(120, kernel_size=5, strides=1, activation='relu', padding='valid'),
        keras.layers.Flatten(),
        keras.layers.Dense(84, activation='relu'),
        keras.layers.Dense(10, activation='softmax')
    ])
    lenet.compile(optimizer='adam', loss=keras.losses.sparse_categorical_crossentropy, metrics=['accuracy'])
    return lenet

lenet = build_lenet()
lenet.summary()

# 3. Training the model
hist = lenet.fit(x_train, y_train, epochs=100, validation_data=(x_val, y_val), verbose=1)

# 4. Plotting accuracy and loss
def plot_training_metrics(hist):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

    ax1.plot(hist.history['accuracy'], label='Train')
    ax1.plot(hist.history['val_accuracy'], label='Validation')
    ax1.set_title("Accuracy by LeNet on CIFAR-10 Data")
    ax1.set_ylabel('Accuracy')
    ax1.set_xlabel('Epochs')
    ax1.legend(loc='upper left')

    ax2.plot(hist.history['loss'], label='Train')
    ax2.plot(hist.history['val_loss'], label='Validation')
    ax2.set_title('Loss by LeNet on CIFAR-10 Data')
    ax2.set_ylabel('Loss')
    ax2.set_xlabel('Epochs')
    ax2.legend(loc='upper left')

    plt.tight_layout()
    plt.show()

plot_training_metrics(hist)

# 5. Confusion matrix and test evaluation
y_predictions = lenet.predict(x_test)
y_predictions = np.argmax(y_predictions, axis=1)

def plot_confusion_matrix(y_test, y_predictions, classes_name):
    plt.figure(figsize=(7, 6))
    plt.title('Confusion matrix', fontsize=16)
    plt.imshow(confusion_matrix(y_test, y_predictions))
    plt.xticks(np.arange(10), classes_name, rotation=45, fontsize=12)
    plt.yticks(np.arange(10), classes_name, fontsize=12)
    plt.colorbar()
    plt.show()

plot_confusion_matrix(y_test, y_predictions, classes_name)

print("Test accuracy:", accuracy_score(y_test, y_predictions))

# Assuming classes_name is already defined
L, W = 8, 8
fig, axes = plt.subplots(L, W, figsize=(20, 20))
axes = axes.ravel()

for i in np.arange(0, L * W):  
    axes[i].imshow(x_test[i])
    axes[i].set_title(f"Predicted = {classes_name[y_predictions[i]]}\n Actual  = {classes_name[y_test[i]]}")
    axes[i].axis('off')

plt.subplots_adjust(wspace=1)
plt.show()

print(hist.history.keys())

def plot_training_metrics(hist):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

    ax1.plot(hist.history['acc'], label='Train')  # Update this line
    ax1.plot(hist.history['val_acc'], label='Validation')  # Update this line
    ax1.set_title("Accuracy by LeNet on CIFAR-10 Data")
    ax1.set_ylabel('Accuracy')
    ax1.set_xlabel('Epochs')
    ax1.legend(loc='upper left')

    ax2.plot(hist.history['loss'], label='Train')
    ax2.plot(hist.history['val_loss'], label='Validation')
    ax2.set_title('Loss by LeNet on CIFAR-10 Data')
    ax2.set_ylabel('Loss')
    ax2.set_xlabel('Epochs')
    ax2.legend(loc='upper left')

    plt.tight_layout()
    plt.show()
```

# AlexNet

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
from sklearn.metrics import confusion_matrix, accuracy_score

# 1. Data Splitting
x_train_full, x_test, y_train_full, y_test = x_train[:45000], x_train[45000:], y_train[:45000], y_train[45000:]
x_train, x_val, y_train, y_val = x_train_full[:35000], x_train_full[35000:], y_train_full[:35000], y_train_full[35000:]

# 2. Define the AlexNet Model
def build_alexnet(input_shape=(32,32,3)):
    AlexNet = keras.models.Sequential()

    # Convolutional Layers
    AlexNet.add(keras.layers.Conv2D(filters=16, kernel_size=(3,3), strides=(4,4), input_shape=input_shape, activation='relu'))
    AlexNet.add(keras.layers.MaxPooling2D(pool_size=(2,2), strides=(2,2)))
    AlexNet.add(keras.layers.Conv2D(60, (5,5), padding='same', activation='relu'))
    AlexNet.add(keras.layers.MaxPooling2D(pool_size=(2,2), strides=(2,2)))
    AlexNet.add(keras.layers.Conv2D(60, (3,3), padding='same', activation='relu'))
    AlexNet.add(keras.layers.Conv2D(30, (3,3), padding='same', activation='relu'))
    AlexNet.add(keras.layers.Conv2D(20, (3,3), padding='same', activation='relu'))
    AlexNet.add(keras.layers.MaxPooling2D(pool_size=(2,2), strides=(2,2)))
    
    # Flatten and Dense Layers
    AlexNet.add(keras.layers.Flatten())
    AlexNet.add(keras.layers.Dense(200, activation='relu'))
    AlexNet.add(keras.layers.Dropout(0.1))
    AlexNet.add(keras.layers.Dense(200, activation='relu'))
    AlexNet.add(keras.layers.Dropout(0.1))
    AlexNet.add(keras.layers.Dense(10, activation='softmax'))

    AlexNet.compile(optimizer='SGD', loss=keras.losses.sparse_categorical_crossentropy, metrics=['accuracy'])
    return AlexNet

AlexNet = build_alexnet()
AlexNet.summary()

# 3. Train the model
history1 = AlexNet.fit(x_train, y_train, epochs=100, validation_data=(x_val, y_val), verbose=1)

# 4. Plotting and Evaluating
def plot_metrics(history):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

    ax1.plot(history.history['accuracy'], label='Train')
    ax1.plot(history.history['val_accuracy'], label='Validation')
    ax1.set_title("Accuracy by AlexNet on CIFAR-10 Data")
    ax1.set_ylabel('Accuracy')
    ax1.set_xlabel('Epochs')
    ax1.legend(loc='upper left')

    ax2.plot(history.history['loss'], label='Train')
    ax2.plot(history.history['val_loss'], label='Validation')
    ax2.set_title('Loss by AlexNet on CIFAR-10 Data')
    ax2.set_ylabel('Loss')
    ax2.set_xlabel('Epochs')
    ax2.legend(loc='upper left')

    plt.tight_layout()
    plt.show()

plot_metrics(history1)

y_predictions1 = AlexNet.predict(x_test)
y_predictions1 = np.argmax(y_predictions1, axis=1)

def plot_confusion_matrix(y_test, y_predictions, classes_name):
    plt.figure(figsize=(7, 6))
    plt.title('Confusion matrix of AlexNet Model', fontsize=16)
    plt.imshow(confusion_matrix(y_test, y_predictions))
    plt.xticks(np.arange(10), classes_name, rotation=45, fontsize=12)
    plt.yticks(np.arange(10), classes_name, fontsize=12)
    plt.colorbar()
    plt.show()

plot_confusion_matrix(y_test, y_predictions1, classes_name)

print("Test accuracy by AlexNet:", accuracy_score(y_test, y_predictions1))
```

# VGG Net

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow import keras
from sklearn.metrics import confusion_matrix, accuracy_score

# Load and preprocess CIFAR-10 dataset
from tensorflow.keras.datasets import cifar10
(x_full, y_full), (x_test, y_test) = cifar10.load_data()
x_full = x_full.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

# 1. Data Splitting
x_train = x_full[:35000]
y_train = y_full[:35000]
x_val = x_full[35000:50000]
y_val = y_full[35000:50000]

# 2. Define VGGNet-inspired model
def build_vggnet():
    VGGNet = keras.models.Sequential()

    # Convolutional Layers
    VGGNet.add(keras.layers.Conv2D(32, (3, 3), activation='relu', padding='same', input_shape=(32, 32, 3)))
    VGGNet.add(keras.layers.Conv2D(32, (3, 3), activation='relu', padding='same'))
    VGGNet.add(keras.layers.MaxPooling2D((2, 2)))

    VGGNet.add(keras.layers.Conv2D(64, (3, 3), activation='relu', padding='same'))
    VGGNet.add(keras.layers.Conv2D(64, (3, 3), activation='relu', padding='same'))
    VGGNet.add(keras.layers.MaxPooling2D((2, 2)))

    # Dense Layers
    VGGNet.add(keras.layers.Flatten())
    VGGNet.add(keras.layers.Dense(512, activation='relu'))
    VGGNet.add(keras.layers.Dropout(0.5))
    VGGNet.add(keras.layers.Dense(10, activation='softmax'))

    VGGNet.compile(optimizer=keras.optimizers.SGD(lr=0.001, momentum=0.9), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return VGGNet

VGGNet = build_vggnet()
VGGNet.summary()

# 3. Train the model
history2 = VGGNet.fit(x_train, y_train, epochs=100, validation_data=(x_val, y_val), verbose=1)

# 4. Plotting and Evaluating
def plot_metrics(history, title_prefix):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

    ax1.plot(history.history['accuracy'], label='Train')
    ax1.plot(history.history['val_accuracy'], label='Validation')
    ax1.set_title(f"Accuracy by {title_prefix} on CIFAR-10 Data")
    ax1.set_ylabel('Accuracy')
    ax1.set_xlabel('Epochs')
    ax1.legend(loc='upper left')

    ax2.plot(history.history['loss'], label='Train')
    ax2.plot(history.history['val_loss'], label='Validation')
    ax2.set_title(f'Loss by {title_prefix} on CIFAR-10 Data')
    ax2.set_ylabel('Loss')
    ax2.set_xlabel('Epochs')
    ax2.legend(loc='upper left')

    plt.tight_layout()
    plt.show()

plot_metrics(history2, "VGGNet")

y_predictions2 = VGGNet.predict(x_test)
y_predictions2 = np.argmax(y_predictions2, axis=1)

classes_name = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']

def plot_confusion_matrix(y_true, y_pred, classes):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=classes, yticklabels=classes)
    plt.xlabel('Predicted labels')
    plt.ylabel('True labels')
    plt.title('Confusion matrix of VGGNet Model')
    plt.show()

plot_confusion_matrix(y_test, y_predictions2, classes_name)
print("Test accuracy by VGGNet:", accuracy_score(y_test, y_predictions2))
```
